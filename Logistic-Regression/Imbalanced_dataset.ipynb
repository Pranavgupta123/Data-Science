```python
import pandas as pd
import numpy as np
```

# Creating Imbalanced Dataset

```python
df1=pd.DataFrame({
    'c1':np.random.normal(loc=0,scale=1,size=900),
    'c2':np.random.normal(loc=0,scale=1,size=900),
    'target':[0]*900
})

df2=pd.DataFrame({
    'c1':np.random.normal(loc=0,scale=1,size=100),
    'c2':np.random.normal(loc=0,scale=1,size=100),
    'target':[1]*100
})

df=pd.concat([df1,df2]).reset_index(drop=True)
df
```

```python
from sklearn.utils import resample

df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
df_upsampled_minority=resample(df_minority,replace=True,n_samples=len(df_majority),random_state=42)

df=pd.concat([df_upsampled_minority,df_majority])
```
